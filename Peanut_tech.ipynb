# Modelagem: Regresso Linear Multipla - STEPWISE


```python
X = dados.loc[:, 'temp_SEP':] #features = variaveis independentes = Xs
y = dados['produtividade'] # target = variavel (is) dependentes = ys
```

```python
y.head()
```

```python
X.head()
```

```python
# valores padronizados
from sklearn.preprocessing import StandardScaler
```

```python
escala = StandardScaler().fit(X)
```

```python
Xpad = escala.transform(X)
```

```python
Xpad
```

```python
Xpad = pd.DataFrame(Xpad, columns=[dados.columns[4:]])
```

```python
#Xpad
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#!pip install mlxtend
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
```

```python
modelo = LinearRegression()
sfs_problema = sfs(modelo, k_features = 20, forward = True, scoring = 'r2', cv = 3) #k_features => reduo de 60 para 20 variaveis
```

```python
sfs_problema = sfs_problema.fit(Xpad, y)
```

```python
sfs_problema
```

```python
tabela = pd.DataFrame.from_dict(sfs_problema.get_metric_dict()).T
```

```python
tabela
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
```

```python
fig1 = plot_sfs(sfs_problema.get_metric_dict(), kind = 'std_dev')
plt.grid()
```

```python
tabela.loc[10,'feature_names']
```

```python
import statsmodels.api as sm
```

```python
xobs = Xpad[['ue_ANN', 'tmax_SEP', 'tmax_DEC', 'tamplitude_NOV', 'Prec_SEP',
              'Prec_DEC', 'Qg_SEP', 'Qg_DEC', 'Qg_ANN', 'ETP_OCT']]
```

```python
yobs = y
```

```python
xobs = sm.add_constant(xobs)
```

```python
xobs.head()
```

```python
resultado = sm.OLS(yobs,xobs).fit()
```

```python
resultado.summary()
```

# 3) Modelagem: Redes Neurais Artificiais - Multilayer Perceptron

```python
Xpad.head()
```

```python
y.head()
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
modelo = MLPRegressor(hidden_layer_sizes=(5,5) , # duas camadas escondidas com 5 neurnios em cada camada
                      learning_rate='adaptive',
                      learning_rate_init=0.05,
                      solver='lbfgs',
                      random_state=1,
                      max_iter=1000)
```

```python
modelo.fit(Xpad, y)
```

```python
yest=modelo.predict(Xpad)
```

```python
dadosgraf = pd.DataFrame({'yobs':y, 'yest':yest})
```

```python
dadosgraf.head()
```

```python
sns.lmplot(x='yobs', y='yest', data=dadosgraf)
```

```python
import sklearn.metrics as me
r2 = me.r2_score(dadosgraf['yobs'], dadosgraf['yest'])
```

```python
r2
```

```python
rmse = me.mean_squared_error(dadosgraf['yobs'], dadosgraf['yest'])
```
